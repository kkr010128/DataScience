## 시도별 데이터

```python
import pandas as pd
a = pd.read_csv("DataScience/data/시도별_인구.csv")
b = pd.read_csv("DataScience/data/시도별_외국인인구.csv")
x = pd.merge(a, b, left_on="시도명", right_on="시도")
x
x["여자인구비"] = x["여자"]/x["총인구"]*100
x["외국인비"] = x["외국인"]/x["총인구"]*100
y = x[['시도명', '총인구', '가구수', '여자인구비', '외국인비']]
y = y.sort_values('외국인비', ascending=False, ignore_index=True)
y
```

```python
import pandas as pd

# 데이터 로드
data = pd.read_csv('DataScience/data/Most_Popular_Programming_Languages.csv')

# 'Date' 열을 제외하고 각 프로그래밍 언어의 평균 점수를 계산한 후, 내림차순으로 정렬
mean_scores = data.drop(columns=['Date']).mean().sort_values(ascending=False)

# 언어, 점수, 순위로 구성된 새로운 DataFrame 생성
mean_scores_df = pd.DataFrame({
    'Language': mean_scores.index,
    'Score': mean_scores.values,
    '순위': range(1, len(mean_scores) + 1)
})

# 인덱스를 재설정하여 최종 결과 출력
mean_scores_df = mean_scores_df.reset_index(drop=True)
mean_top = mean_scores_df.reset_index(drop=True).head(10)

# 결과 출력
mean_scores_df
```

```python
# 데이터 로드
data = pd.read_csv('DataScience/data/Most_Popular_Programming_Languages.csv')

# 'Date' 열을 datetime 형식으로 변환
data['Date'] = pd.to_datetime(data['Date'])

# 2022년 데이터만 필터링
data_2022 = data[data['Date'].dt.year == 2022]

# 'Date' 열을 제외하고 각 프로그래밍 언어의 평균 점수를 계산한 후, 내림차순으로 정렬
mean_scores_2022 = data_2022.drop(columns=['Date']).mean().sort_values(ascending=False)

# 언어, 점수, 순위로 구성된 새로운 DataFrame 생성
mean_scores_2022_df = pd.DataFrame({
    'Language': mean_scores_2022.index,
    'Score': mean_scores_2022.values,
    '순위': range(1, len(mean_scores_2022) + 1)
})

# 인덱스를 재설정하여 최종 결과 출력 (상위 10개만)
mean_scores_2022_df = mean_scores_2022_df.reset_index(drop=True).head(10)

# 결과 출력
mean_scores_2022_df
```
